# Background


## Basic Idea

Wind farm modelling is typically approached from the turbine level up, but what if we approached it from the farm level, or perhaps from the turbine level including all wind direction? The benefit of that would be potentially reducing running many calculations for all wind directions. I think we may be able to reduce to running many calculations a single time and then superimposing the wind distribution in a single loop over all directions.

## Radial Basis Function Types

#### Gaussian:

\begin{align}
 \varphi =& e^{-(\epsilon r)^2}\\
 r =& sqrt(x^2+y^2) \\
 \varphi =& e^{-\big(\epsilon \sqrt{x^2+y^2}\big)^2}
\end{align}

```julia
# define function
g(x,y) = exp.(-(sqrt.(x.^2 .+ y.^2)).^2)
# define domain
x = collect(-5:0.01:5)
y = collect(-5:0.01:5)
# calculate result
gout = g(x,y)
r = sqrt.(x.^2 .+ y.^2);
```

```julia
# plot in 2D
plot(r,gout)
```

```julia
# plot in 3D
plot(x,y,g,st=:surface,color=:Blues)
```

#### Multiquadratic:

\begin{equation}
    \varphi = \sqrt{1+(\epsilon r)^2}
\end{equation}

The multiquadratic form is similar to wind turbine wake deficit if general look, and could possilby be changed to fit the form more exactly.

```julia
# define function
m(x,y;eps=1.0) = sqrt.(1 .+ (eps*sqrt.(x.^2 .+ y.^2)).^2)
# calculate result
mout = m(x,y);
```

```julia
# plot in 2D
plot(r,mout)
```

```julia
# plot in 3D
plot(x,y,m,st=:surface,color=:Blues)
```

## Combining with Windrose function

Using a Rayleigh distribution

\begin{equation}
    f(x,\sigma) = \frac{x}{\sigma^2} e^{-x^2/(2\sigma^2)}, \quad x\ge0
\end{equation}

```julia
rf(angle;sigma=1) = (angle./sigma^2).*exp(-angle.^2 ./(2*sigma^2));
angle = 0:0.1:2*pi
```

```julia
# plot windrose probability
plot(angle,rf,proj=:polar)
```

```julia
# full function
function combined_function(x, y)
    r = sqrt.(x.^2 .+ y.^2)
    gout = g(x,y)
    angle = 0.0
    if x >= 0 && y >= 0 
        angle = acos.(x./r)
    elseif x < 0 && y >= 0
        angle = pi - acos.(abs.(x)./r)
    elseif x < 0 && y < 0
        angle = acos.(abs.(x)./r) + pi
    elseif x >= 0 && y < 0
        angle = 2*pi - acos.(abs.(x)./r)
    end
#     println(angle*180.0/pi)
    scale = rf(abs(angle))
    return scale*gout
end
```

```julia
# plot in 2D
plot(x,y,combined_function,st=:surface,color=:Blues,xlabel="x",ylabel="y",camera=(90,90))
```
